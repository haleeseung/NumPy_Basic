# Section 6. N차원 배열의 병합

### 6-1. 배열에 원소 추가 및 삭제

```python
# Python list
arr = [1, 2, 3, 4, 5, 6, 7, 8]
arr.insert(2, 50)
print(arr)

# 1차원 배열
arr = np.arange(1, 9)
arr = np.insert(arr, 2, 50)
print(arr)
```

```python
# 2차원 배열
arr = np.arange(1, 13).reshape(3, 4)
arr = np.insert(arr, 2, 50)
print(arr)
```

```python
# delete
arr = np.arange(1, 13).reshape(3, 4)
print(arr)

arr = np.delete(arr, 2)
print(arr)
```

### 6-4. 배열 간의 병합

```python
# append()
arr1 = np.arange(1, 13).reshape(3, 4)
arr2 = np.arange(13, 25).reshape(3, 4)

print(arr1)
print(arr2, end="\n\n")

# axis = 0
arr3 = np.append(arr1, arr2, axis=0)
print(arr3, end="\n\n")

# axis = 1
arr3 = np.append(arr1, arr2, axis=1)
print(arr3, end="\n\n")

# axis가 없으면
arr3 = np.append(arr1, arr2)
print(arr3, end="\n\n")
```

```python
# vstack(), hstack()
arr1 = np.arange(1, 7).reshape(2, 3)
arr2 = np.arange(7, 13).reshape(2, 3)

arr3 = np.vstack((arr1, arr2))
print(arr3)

arr4 = np.hstack((arr1, arr2))
print(arr4)
```

```python
# concatenate()
arr1 = np.arange(1, 7).reshape(2, 3)
arr2 = np.arange(7, 13).reshape(2, 3)

arr3 = np.concatenate([arr1, arr2], axis=0)
print(arr3)

arr4 = np.concatenate([arr1, arr2], axis=1)
print(arr4)
```

### 6-3. 배열 분할

```python
# 2차원 배열
arr = np.arange(1, 13).reshape(3, 4)
print(arr)

# vsplit() => axis=0
arr_vsplit = np.vsplit(arr, 3)
print(arr_vsplit)

# hsplit() => axis=1
arr_hsplit = np.hsplit(arr, 2)
print(arr_hsplit)
```

```python
# 3차원 배열
arr = np.random.randint(0, 10, (4, 6, 8))
print(arr)
```

```python
arr_vsplit = np.vsplit(arr, 2)
print(arr_vsplit)
```

```python
arr_hsplit = np.hsplit(arr, 2)
print(arr_hsplit)
```
