```python
import numpy as np
import matplotlib.pyplot as plt
```

# Section 1. N차원 배열 생성

### 1-1. N차원 배열(ndarray) 생성하기

```python
# 1차원 배열
arr = np.array([1, 2, 3])
print(arr)

# 2차원 배열
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
print(arr)
```

```python
type([1, 2, 3])
type(arr)
```

```python
tpl = (4, 5, 6)
arr = np.array(tpl)
print(arr)

lst = [1, 2, 3]
arr = np.array(lst)
print(arr)

lst2 = [[1, 2, 3], [4, 5, 6]]
arr2 = np.array(lst2)
print(arr2)
```

```python
# shape
arr1 = np.array([1, 2, 3])
arr2 = np.array([[1, 2, 3], [4, 5, 6]])

print(arr1.shape, arr2.shape)

# ndim
print(arr1.ndim, arr2.ndim)

# size (원소의 개수)
print(arr1.size, arr2.size)
```

### 1-2. N차원 배열의 데이터 타입

```python
arr = np.array([0, 1, 2, 3])
print(arr, arr.dtype)

arr = arr.astype(np.float32)
print(arr, arr.dtype)
```

```python
# 데이터 타입이 혼재하는 경우
arr = np.array([1, 2, 3.4, "64, 문자열"], dtype=int)
print(arr, arr.dtype)
```

### 1-3. 정해진 형식의 N차원 배열 생성하기

```python
# np.zeros()
arr = np.zeros([2, 2])
print(arr)

# np.ones()
arr = np.ones([3, 5])
print(arr)

# np.full()
arr = np.full((2, 3), 5)
print(arr)

# np.eye()
arr = np.eye(3)
print(arr)
```

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

# np.zeros_like()
arr_z = np.zeros_like(arr)
print(arr_z)

# np.ones_like()
arr_o = np.ones_like(arr)
print(arr_o)

# np.full_like()
arr_f = np.full_like(arr, 9)
print(arr_f)
```

### 1-4. 특정 범위의 값을 가지는 N차원 배열 생성하기

```python
lst = list(range(0, 9, 2))
print(lst)
```

```python
arr = np.arange(9)
print(arr)
arr = np.arange(stop=9, start=0, step=2) # (0, 9, 2)
print(arr)
```

```python
# np.linspace()
arr = np.linspace(0, 100, 250)
print(arr)
```

```python
# np.logspace()

arr = np.linspace(1, 10, 10)
print(arr, end="\n\n")

arr = np.logspace(1, 10, 10, base=2)
print(arr)

arr = np.logspace(1, 10, 10)
print(arr)
```

### 1-5. 난수로 N차원 배열 생성하기

```python
# np.random.normal()

arr = np.random.normal(loc = 0, scale = 1, size = (2, 3))
print(arr)
```

```python
arr = np.random.normal(0, 1, 1000)
plt.hist(arr, bins=100)
plt.show()
```

```python
# np.random.rand()
arr = np.random.rand(1000)
plt.hist(arr, bins=100)
plt.show()
```

```python
# np.random.randn()
arr = np.random.randn(1000)
plt.hist(arr, bins=100)
plt.show()
```

```python
# np.random.randint()
arr = np.random.randint(low=1, high=5, size=(3, 4))
print(arr)
```

```python
arr = np.random.randint(100, 200, 1000)
plt.hist(arr, bins=100)
plt.show()
```

```python
# rand(), randn(), randint()
```

### 1-6. 시드(Seed)값을 통한 난수 생성 제어

```python
arr = np.random.rand(10)
print("난수 발생1 \n", arr)

arr = np.random.rand(10)
print("난수 발생2 \n", arr)
```

```python
np.random.seed(10)
arr = np.random.rand(10)
print("난수 발생1 \n", arr)

np.random.seed(10)
arr = np.random.rand(10)
print("난수 발생2 \n", arr)
```
