# Section 4. N차원 배열 정렬

### 4-1. 1차원 배열의 정렬

```python
arr = np.random.randint(10, size=10)
print(arr)
```

```python
print(np.sort(arr))
print(np.sort(arr)[::-1])
```

```python
arr = np.sort(arr)
print(arr)
arr.sort()
print(arr)
```

### 4-2. 2차원 배열의 정렬

```python
arr = np.random.randint(15, size=(3, 4))
print(arr)
```

```python
print(np.sort(arr, axis=None)) # axis = None으로 하면 1차원 배열로 변환시켜 오름차순 정렬시킴
```

```python
# argsort()
print(np.sort(arr, axis=0))
print(np.argsort(arr, axis=0)) # 그 값의 index에 해당되는 부분을 정렬시킴.
```
